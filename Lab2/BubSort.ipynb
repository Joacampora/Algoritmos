```python
import matplotlib.pyplot as plt
import random
def perms(n):
    v=[]
    for i in range(n):
        v.append(i+1)
    for i in range(n-1):
        j = random.randint(i, n-1)
        aux = v[i]
        v[i] = v[j]
        v[j] = aux
    return v

def bsortPasos(a):
    v = []
    for i in range(len(a)):
        v.append(a[i])

    Pasos = 0
    
    for i in range(len(v)-1):
        for j in range(len(v)-1):
            if v[j] > v[j+1]:
                temp = v[j]
                v[j] = v[j+1]
                v[j+1] = temp
                Pasos = Pasos + 3
            Pasos = Pasos + 1
    Pasos = Pasos + 1
    
    return Pasos


n = 10
runs = 100
pasos=[]
sum = 0
min = n**3
max = 0
for i in range(runs):
    t = bsortPasos(perms(n))    
    pasos.append(t)
    if t < min :
       min = t
    if t > max :
       max = t
    sum = sum + t
print pasos
print 'Mejor Teorico: ' + str(5*n - 4)
print 'Peor Teorico: ' + str(((n+1)*n)/2)
print 'Promedio Teorico: ' + str(((n+1)*n)/4)
print 'Mejor: ' + str(min)
print 'Peor: ' + str(max)
print 'Promedio: ' + str(sum/runs)
plt.hist(pasos, normed=True, bins=runs);
```

```python
import itertools
import random
def perms(n):
    perm = []
    for i in range(n):
        perm.append("")
        perm[i] = i+1
    for i in range(n-1):
        x = random.randint(i, n-1)
        temp = perm[i]
        perm[i] = perm[x]
        perm[x] = temp
    return perm

def bsortPasos(a):
    v = []
    for i in range(len(a)):
        v.append(a[i])

    Pasos = 0
    
    for i in range(len(v)-1):
        for j in range(len(v)-1):
            if v[j] > v[j+1]:
                temp = v[j]
                v[j] = v[j+1]
                v[j+1] = temp
                Pasos = Pasos + 3
            Pasos = Pasos + 1
    Pasos = Pasos + 1
    
    return Pasos

n = 8
perm = []
count = 0
sum = 0
best = n ** 3
worst = 0
perm = perms(n)
print perm
pasos=[]
for i in itertools.permutations(perm):
    x = bsortPasos(i)
    pasos.append(x)
    if x < best:
        best = x
    if x > worst:
        worst = x
    sum += x
    count += 1

print pasos
print 'Mejor Teorico, ' + str(5*n - 4) 
print 'Peor teorico,' + str((3.0/2.0)*n**2 + (7.0/2.0)*n - 4)
print 'Promedio teorico,' + str((3.0/4.0)*n**2 + (17.0/4.0)*n - 4)
print 'Mejor: ' + str(best)
print 'Peor: ' + str(worst)
print 'Promedio: ' + str(sum / count)
plt.hist(pasos, normed=True, bins=runs);
```
