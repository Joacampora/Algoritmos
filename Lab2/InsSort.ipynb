```python
import matplotlib.pyplot as plt
import random
def perms(n):
    v=[]
    for i in range(n):
        v.append(i+1)
    for i in range(n-1):
        j = random.randint(i, n-1)
        aux = v[i]
        v[i] = v[j]
        v[j] = aux
    return v

def isortSteps(a):
    #print a
    v = []
    for i in range(len(a)):
        v.append(a[i])
        
    steps = 0
    for i in range(1,len(v)):
        x = v[i]
        j = i-1
        while (j > -1) and (v[j] > x):
            v[j+1] = v[j]
            j = j -1
            steps = steps + 3
        steps = steps + 1
        v[j+1] = x
        steps = steps + 4
    steps = steps + 1
    #print v
    return steps


n = 10
runs = 100
pasos=[]
sum = 0
min = n**3
max = 0
for i in range(runs):
    t = isortSteps(perms(n))    
    pasos.append(t)
    if t < min :
       min = t
    if t > max :
       max = t
    sum = sum + t
print pasos
print 'Mejor Teorico, ' + str(5*n - 4) 
print 'Peor teorico,' + str((3.0/2.0)*n**2 + (7.0/2.0)*n - 4)
print 'Promedio teorico,' + str((3.0/4.0)*n**2 + (17.0/4.0)*n - 4)
print 'Mejor, ' + str(min)
print 'Peor,' + str(max)
print 'Promedio,' + str(sum/runs)
plt.hist(pasos, normed=True, bins=runs);
```

```python
import itertools
import random
def perms(n):
    perm = []
    for i in range(n):
        perm.append("")
        perm[i] = i+1
    for i in range(n-1):
        x = random.randint(i, n-1)
        temp = perm[i]
        perm[i] = perm[x]
        perm[x] = temp
    return perm

def instrucciones(x):
    perm = []
    for i in range(len(x)):
        perm.append([])
        perm[i] = x[i]

    instructions = 0
    for i in range(1, len(perm)):
        x = perm[i]
        j = i - 1
        while (j > -1) and (perm[j] > x):
            perm[j + 1] = perm[j]
            j = j - 1
            instructions += 3
        instructions += 1
        perm[j + 1] = x
        instructions += 4
    instructions += 1
    return instructions

n = 9
perm = []
count = 0
sum = 0
best = n ** 3
worst = 0
perm = perms(n)
print perm
pasos=[]
for i in itertools.permutations(perm):
    x = instrucciones(i)
    pasos.append(x)
    if x < best:
        best = x
    if x > worst:
        worst = x
    sum += x
    count += 1

print pasos
print 'Mejor Teorico, ' + str(5*n - 4) 
print 'Peor teorico,' + str((3.0/2.0)*n**2 + (7.0/2.0)*n - 4)
print 'Promedio teorico,' + str((3.0/4.0)*n**2 + (17.0/4.0)*n - 4)
print 'Mejor: ' + str(best)
print 'Peor: ' + str(worst)
print 'Promedio: ' + str(sum / count)
plt.hist(pasos, normed=True, bins=runs);
```
