```python
import time
import matplotlib.pyplot as plt
x = []
def fibo(num):
    fib=0
    aux1=1
    aux2=1
    i=2
    ti= time.clock()
    if(num==1 or num==2):
        fib=1
    else:
        while(i<num):
            
            fib=aux1+aux2
            aux2=aux1
            aux1=fib
            i=i+1
    return time.clock()-ti
for i in range(2000):
    x += [fibo(i)]
plt.plot(x, linewidth=1);
plt.xlabel('Termino') ;
plt.ylabel('Tiempo');
```
