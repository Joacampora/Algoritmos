```python
from random import randint
import matplotlib.pyplot as plt
import time
import numpy
t=100


times_prom=[]

for h in range(1,t):
    ins= 2*(h**3) - h**2
    times=[]
    result = [[0 for x in range(h)] for y in range(h)]
    m1 = [[randint(1, 3) for x in range(h)] for y in range(h)]
    m2 = [[randint(1, 3) for x in range(h)] for y in range(h)]
    for l in range(10):
        init = time.time()
        for i in range(len(m1)):
            for j in range(len(m2[0])):
                for k in range(len(m2)):
                    result[i][j] += m1[i][k] * m2[k][j]
        times.append((time.time()-init)/ins)
    times_prom.append(sum(times)/10)
            
print len(times)
print len(times_prom)
plt.plot(times_prom, linewidth=1);
```
